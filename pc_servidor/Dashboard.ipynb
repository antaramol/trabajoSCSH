```python
#%%
import pymongo as mongo

#connect to the database SistCiber and the collection Accel
db = mongo.MongoClient("mongodb://localhost:27017/")["SistCiber"]
collection = db["Accel"]
#find all the documents in the collection
documents = collection.find()
data_64 = []
data_1024 = []

for document in documents:
    #save each document in a different list
    #remove the _id field
    document.pop("_id")
    if document["muestras"] == "64":
        document.pop("muestras")
        data_64.append(document)
    else:
        document.pop("muestras")
        data_1024.append(document)
```

```python
import matplotlib.pyplot as plt
import numpy as np
collection = db["Temperature"]
documents = collection.find()
Temperature = []
Datetime = []

for document in documents:
    document.pop("_id")
    Datetime.append(document["datetime"])
    Temperature.append(document["temperature"])

#plot the temperature vs time
plt.figure(figsize=(10,10))
plt.title("Temperature")
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.plot(Datetime, Temperature)
plt.xticks(rotation=30)
plt.xticks(np.arange(0, len(Datetime), 10))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
collection = db["Humidity"]
documents = collection.find()
Humidity = []
Datetime = []

for document in documents:
    document.pop("_id")
    Datetime.append(document["datetime"])
    Humidity.append(document["humidity"])

Humidity = np.array(Humidity, dtype=int)
#plot the Humidity vs time
plt.figure(figsize=(10,10))
plt.title("Humidity")
plt.xlabel("Time")
plt.ylabel("Humidity(%)")
plt.plot(Datetime, Humidity)
plt.xticks(rotation=30)
plt.xticks(np.arange(0, len(Datetime), 10))
plt.show()
```
